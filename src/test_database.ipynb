# Cycling Data Database Testing

This notebook provides SQL queries to test and explore the cycling data database after scraping.


```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('../data/cycling_data.db')

print("Connected to cycling database successfully!")
```

## Database Schema Overview

```python
# Check all tables in the database
tables_query = """

  SELECT
    *
  FROM results res
  JOIN stages s ON res.stage_id = s.id
  JOIN races r ON s.race_id = r.id
  WHERE r.year = 2024
    AND res.rider_url LIKE '%tadej-pogacar%'
"""

tables = pd.read_sql_query(tables_query, conn)
print("Output:")
print(tables)
```

```python
# Close the database connection when done
conn.close()
print("Database connection closed.")
```
